```python
"""The algorithm maintains a map (num_map) where each number in the array is mapped to its index. It iterates through the
   array and for each number, it computes the complement (the difference between the target and the current number). If the
  complement exists in the map, it means that the two numbers add up to the target, so it returns the indices. Otherwise,
  it adds the number to the map and continues. If no solution is found, it returns an empty list (or you can handle it 
  differently based on your needs)."""
```

```python
#<aside>
ðŸ’¡ **Q2.** Given an integer array nums and an integer val, remove all occurrences of val in nums in-place. The order of 
         the elements may be changed. Then return the number of elements in nums which are not equal to val.

   Consider the number of elements in nums which are not equal to val be k, to get accepted, you need to do the following 
    things:

- Change the array nums such that the first k elements of nums contain the elements which are not equal to val. The 
   remaining elements of nums are not important as well as the size of nums.
- Return k.

**Example :**
Input: nums = [3,2,2,3], val = 3
Output: 2, nums = [2,2,_*,_*]

**Explanation:** Your function should return k = 2, with the first two elements of nums being 2. It does not matter what 
you leave beyond the returned k (hence they are underscores)[

</aside>
    
"""To remove all occurrences of a given value from an integer array nums in-place and return the number of elements that 
   are not equal to the value, you can use a two-pointer approach. Here's the solution in Python:"""

def removeElement(nums, val):
    k = 0  # Number of elements not equal to val

    # Iterate through the array
    for i in range(len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]
            k += 1

    return k

# Example usage
nums = [3, 2, 2, 3]
val = 3
k = removeElement(nums, val)
print(k)  # Output: 2
print(nums)  # Output: [2, 2, _, _]
```

```python
"""The algorithm uses two pointers: i to iterate through the array and k to keep track of the next index to store elements 
   not equal to val. Whenever nums[i] is not equal to val, it means it's one of the elements we want to keep. We store it at
   index k and increment k by 1. After iterating through the entire array, the first k elements will contain the elements not
   equal to val. The remaining elements are not important and can be ignored. The algorithm returns k, which represents the
   number of elements not equal to val.
   
   
   Note: In the example output provided, underscores (_) are used to represent the remaining elements of nums beyond the 
   returned k. However, in reality, the values of those elements remain unchanged, and only the first k elements are
   meaningful."""
```

```python
#<aside>
ðŸ’¡ **Q3.** Given a sorted array of distinct integers and a target value, return the index if the target is found. If not,
    return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

**Example 1:**
Input: nums = [1,3,5,6], target = 5

Output: 2

</aside>

"""To solve this problem with O(log n) runtime complexity, we can use the binary search algorithm. Here's the algorithm 
   to find the index of the target value or the index where it would be inserted:
   
     1. Initialize two pointers, left and right, pointing to the start and end of the array respectively.
     
     2. While left is less than or equal to right, do:
        a. Calculate the middle index as mid = (left + right) / 2.
        b. If the value at the middle index is equal to the target, return the middle index.
        c. If the value at the middle index is greater than the target, update right = mid - 1.
        d. If the value at the middle index is less than the target, update left = mid + 1.
        
     3. If the target is not found in the array, the value of left will represent the index where it should be inserted.
       Return left.
       
 Here's the implementation in Python:"""

def searchInsert(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left


```

```python
"""Let's test the example given:
"""
nums = [1, 3, 5, 6]
target = 5

print(searchInsert(nums, target))
```

```python
"""The target value 5 is found at index 2 in the given array."""
```

```python
def merge(nums1, m, nums2, n):
    p1 = m - 1
    p2 = n - 1
    p = m + n - 1
    
    while p1 >= 0 and p2 >= 0:
        if nums1[p1] > nums2[p2]:
            nums1[p] = nums1[p1]
            p1 -= 1
        else:
            nums1[p] = nums2[p2]
            p2 -= 1
        p -= 1
    
    while p2 >= 0:
        nums1[p] = nums2[p2]
        p2 -= 1
        p -= 1
#You can call the merge function with the given inputs like this:
        
    nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3

merge(nums1, m, nums2, n)
print(nums1)  # Output: [1, 2, 2, 3, 5, 6]
    
```

```python
#<aside>
ðŸ’¡ **Q7.** Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the nonzero
    elements.

Note that you must do this in-place without making a copy of the array.

**Example 1:**
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

</aside>

"""To move all zeros to the end of the integer array nums while maintaining the relative order of the non-zero elements,
   you can use a two-pointer approach."""

#Here's the Python code that implements this approach:

def moveZeroes(nums):
    n = len(nums)
    zero_ptr = 0  # Pointer to keep track of the position to swap zeros
    for i in range(n):
        if nums[i] != 0:
            # Swap the non-zero element with the position pointed by zero_ptr
            nums[i], nums[zero_ptr] = nums[zero_ptr], nums[i]
            zero_ptr += 1

"""In this code, we initialize a variable zero_ptr to keep track of the position where the next zero should be placed. 
   We iterate through the array using the i pointer. If the element at nums[i] is non-zero, we swap it with the element 
   at nums[zero_ptr], and then increment zero_ptr by 1. This ensures that non-zero elements are moved towards the beginning 
   of the array, while zeros are pushed towards the end.

   Using this approach, the code will transform the example input [0,1,0,3,12] into [1,3,12,0,0], as required."""            
```
